```python
import schedule
import time
import datetime
```

```python
def MINUTE():
    TNOW=datetime.datetime.now().replace(microsecond=0)
    print(str(TNOW)+"This prints every minute")
    
def SECONDS():
    TNOW=datetime.datetime.now().replace(microsecond=0)
    print(str(TNOW)+"tHIS PRINTS EVERY 5 SECONDS INTERVAL FOR 3 TIME")
    
schedule.every(1).minutes.do(MINUTE)
schedule.every().minutes.at(":05").do(SECONDS)
schedule.every().minutes.at(":10").do(SECONDS)
schedule.every().minutes.at(":15").do(SECONDS)

while True:
    schedule.run_pending()
    time.sleep(1)
```
